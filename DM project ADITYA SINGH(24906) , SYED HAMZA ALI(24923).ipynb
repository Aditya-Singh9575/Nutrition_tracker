BUT , DATA IS NOT IN PROPER FORMAT IN WHICH WE WANT . SO HERE NOW WE WILL SELECT SOME SPECIFIC  COLUMNS AND ROWS AND THEN MERGE THEM TO CREATE A NEEDED DATASET.

```python
values_to_remove2 = ['Herbs&Spices','VegetableOils',
       'BakingIngredients','AlcoholicDrinks&Beverages', 'Beer',
       'Non-AlcoholicDrinks&Beverages','CannedFruit', 'PotatoProducts','Fish&Seafood','Cheese', 'CreamCheese','SlicedCheese','Beef&Veal','ColdCuts&LunchMeat','Meat', 'Offal&Giblets', 'Pork',
       'Poultry&Fowl', 'Sausage', 'Venison&Game','Pastries','Breads&Rolls', 'Sauces&Dressings','Soda&SoftDrinks', 'Wine',    'Candy&Sweets', 'IceCream', '(Fruit)Juices', 'CerealProducts',
       'Oatmeal,Muesli&Cereals', 'Pasta&Noodles', 'Dishes&Meals', 'Soups',
       'Legumes', 'Nuts&Seeds', 'Oils&Fats', 'Pastries,Breads&Rolls',
       'Spreads']
mask2 = ~df['FoodCategory'].isin(values_to_remove2)
df_filtered2 = df[mask2]
```

```python
CakesandPies_df = df_filtered2[df_filtered2['FoodCategory']=='Cakes&Pies']
```

```python
result = pd.concat([fruits_df, TropicalandExoticFruits_df,Vegetables_df,FastFood_df,Pizza_df,MilkandDairyProducts_df,Yogurt_df,CakesandPies_df], axis=0)
```

```python
result
```

```python
result.info()
```

```python
result['Cals_per100grams'] = result['Cals_per100grams'].str.extract('(\d+)').astype(int)
result['KJ_per100grams'] = result['KJ_per100grams'].str.extract('(\d+)').astype(float)
```

```python
def categorize_calories(calories):
    if calories < 60:
        return 'low_calorie'
    elif 60 <= calories < 150:
        return 'moderate_calorie'
    else:
        return 'high_calorie'
```

```python
result['calorie_count'] = result['Cals_per100grams'].apply(categorize_calories)
```

```python
def categorize_fat_category(category):
    if category in ['Fruits', 'Tropical&ExoticFruits', 'Vegetables']:
        return 'low_fat'
    else:
        return 'high_fat'
```

```python
result['fat-category']= result['FoodCategory'].apply( categorize_fat_category)
```

# now our dataset is ready for training

# now we will train our model

In the first view , it may seem like "we could have also written a simple program by using a threshold value to determine the required category of food" . But here along with the fat-content and calorie-count ,model is also being trained on the basis of food-category (for ex. if there is a food-category that has been placed in  a unhealthy category for most of the time in historical data on which we have trained our model , then if in future there comes some food-item belonging to same category but with  low-calorie(per 100g), then as our model is now trained or have gained experience . so it will still categorize it to unhealthy or high fat .

```python
result = pd.get_dummies(result, columns=['FoodCategory'])
```

```python
result
```

```python
X = result.drop(['FoodItem', 'fat-category', 'calorie_count'], axis=1)    # Features
y_fat = result['fat-category']                                            # Target: fat-category
y_calorie = result['calorie_count']                                       # Target: calorie_count
```

```python
X_train, X_test, y_fat_train, y_fat_test = train_test_split(X, y_fat, test_size=0.3, random_state=42)
X_train, X_test, y_calorie_train, y_calorie_test = train_test_split(X, y_calorie, test_size=0.3, random_state=42)
```

```python
# Initialize and train the decision tree classifier for fat-category

fat_classifier = DecisionTreeClassifier()
fat_classifier.fit(X_train, y_fat_train)

# Initialize and train the decision tree classifier for calorie_count

calorie_classifier = DecisionTreeClassifier()
calorie_classifier.fit(X_train, y_calorie_train)
```

```python
# Predictions

fat_predictions = fat_classifier.predict(X_test)
calorie_predictions = calorie_classifier.predict(X_test)
```

```python
#accuracy

fat_accuracy = accuracy_score(y_fat_test, fat_predictions)
calorie_accuracy = accuracy_score(y_calorie_test, calorie_predictions)
```

```python
print("Accuracy for fat-category classification:", fat_accuracy)
print("Accuracy for calorie_count classification:", calorie_accuracy)
```
